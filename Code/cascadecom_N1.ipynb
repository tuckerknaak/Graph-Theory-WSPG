#### A cascade is defined as a sequence of chain reactions through a network, and a complete cascade results in the entire network being affected by the chain reactions.  A cluster of nodes has a densiy $p$ if each node in the set has at least a $p$ fraction of its neighbors in the cluster.  It can be shown that there exists a threshold $q$, such that a complete cascade cannot occur if a cluster of nodes in the network has a density $p>1-q$.  In this code, we will begin by having the nodes from the Louvain Community with the largest total betweenness centrality fail.  We will then determine whether these failed nodes will cause a complete through the network for different threshold values.
